Gráfico 2 - Gráfico com todos os produtos vendidos por mês - Linha

```python
import matplotlib.pyplot as plt
import numpy as np

def somarMesesProdutos(produto1, produto2, produto3, produto4, produto5, produto6):
    """Soma todas as vendas, de todos os produtos, de cada mês. Assim, cada mês apresenta o total de vendas de todos os produtos."""
    
    listaSomaMeses = []

    for i in range(0, 12):
        listaSomaMeses.append(produto1[i] + produto2[i] + produto3[i] + produto4[i] + produto5[i] + produto6[i])
        
    return listaSomaMeses

resultadoSomaProdutos = somarMesesProdutos(creme_facial, limpeza_facial, pasta_dentaria, sabonete, shampoo, hidratante)

eixo_x_meses = np.arange(12) + 1
eixo_y_soma_produtos =  resultadoSomaProdutos

fig, ax = plt.subplots()

ax.plot(eixo_x_meses, eixo_y_soma_produtos)
ax.set_title('Venda total')
ax.set_xlabel('Meses')
ax.set_ylabel('Quantidade de produtos vendidos')

fig.set_size_inches(20, 8)
fig.show()
```

Gráfico 4 - Histograma de quantidade de meses (y) e faixas de quantidades de produtos vendidos (1000-1999, 2000-2999, ...)

```python
import matplotlib.pyplot as plt
import numpy as np

creme_facial = np.array([ 2500, 2630, 2140, 3400, 3600, 2760, 2980, 3700, 3540, 1990, 2340, 2900 ])
limpeza_facial = np.array([ 1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760 ])
pasta_dentaria = np.array([ 5200, 5100, 4550, 5870, 4560, 4890, 4780, 5860, 6100, 8300, 7300, 7400 ])
sabonete = np.array([ 9200, 6100, 9550, 8870, 7760, 7490, 8980, 9960, 8100, 10300, 13300, 14400 ])
shampoo = np.array([ 1200, 2100, 3550, 1870, 1560, 1890, 1780, 2860, 2100, 2300, 2400, 1800 ])
hidratante = np.array([ 1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760 ])

def somarMesesProdutos(produto1, produto2, produto3, produto4, produto5, produto6):
    """Soma todas as vendas, de todos os produtos, de cada mês. Assim, cada mês apresenta o total de vendas de todos os produtos."""
    
    listaSomaMeses = []

    for i in range(0, 12):
        listaSomaMeses.append(produto1[i] + produto2[i] + produto3[i] + produto4[i] + produto5[i] + produto6[i])

    return listaSomaMeses

def somarTotalProdutos(produtos):
    """Soma todas as vendas de todos os produtos no ano"""

    somaTotalProdutos = sum(produtos)

    return somaTotalProdutos

def definirFaixaQuantidadeProdutos(somaProdutos):
    """Define a feixa de vendas dos produtos (1000 - 1999, 2000 - 2999,...)"""

    faixasQuantidadesProdutos = []
    
    for i in range(1000, somaProdutos, 1000):
        faixasQuantidadesProdutos.append(i + 999)

    return faixasQuantidadesProdutos

resultadoSomaMesesProdutos = somarMesesProdutos(creme_facial, limpeza_facial, pasta_dentaria, sabonete, shampoo, hidratante)
resultadoSomaTotalProdutos = somarTotalProdutos(resultadoSomaMesesProdutos)
faixasQuantidadesProdutos = definirFaixaQuantidadeProdutos(resultadoSomaTotalProdutos)

eixo_x = faixasQuantidadesProdutos

fig, ax = plt.subplots()

ax.hist(eixo_x, bins = 12, edgecolor = "white", label = 'Montante por mês')
ax.set_title('Histograma - Quantidade de meses e faixas de produtos vendidos')
ax.set_xlabel('Faixa de produtos vendidos')
ax.legend()

fig.set_size_inches(20, 10)
fig.show()
```

Gráfico 5 - Pizza. % da quantidade de produtos vendidos no ano (de cada produto)

```python
import matplotlib.pyplot as plt
import numpy as np

creme_facial = np.array([ 2500, 2630, 2140, 3400, 3600, 2760, 2980, 3700, 3540, 1990, 2340, 2900 ])
limpeza_facial = np.array([ 1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760 ])
pasta_dentaria = np.array([ 5200, 5100, 4550, 5870, 4560, 4890, 4780, 5860, 6100, 8300, 7300, 7400 ])
sabonete = np.array([ 9200, 6100, 9550, 8870, 7760, 7490, 8980, 9960, 8100, 10300, 13300, 14400 ])
shampoo = np.array([ 1200, 2100, 3550, 1870, 1560, 1890, 1780, 2860, 2100, 2300, 2400, 1800 ])
hidratante = np.array([ 1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760 ])

def somarQuantidadeProdutos(produto1, produto2, produto3, produto4, produto5, produto6):
    """Soma a venda de todos os meses de cada produto para encontrar o valor total no ano. Depois, cada total é atribuído a uma posição em uma lista"""
    
    listaSomaQuantidadesProdutos = [sum(produto1), sum(produto2), sum(produto3), sum(produto4), sum(produto5), sum(produto6)]
        
    return listaSomaQuantidadesProdutos

def alterarTamanhoFonte():
    """Autera a fonte globalmente do gráfico"""
    
    plt.rcParams['font.size'] = 18


resultadoSomaProdutos = somarQuantidadeProdutos(creme_facial, limpeza_facial, pasta_dentaria, sabonete, shampoo, hidratante)

eixo_y_soma_produtos =  resultadoSomaProdutos

alterarTamanhoFonte()

fig, ax = plt.subplots()

textos = ['Creme Facial', 'Limpeza Facial', 'Pasta Dentaria', 'Sabonete', 'Shampoo', 'Hidratante']
ax.pie(eixo_y_soma_produtos, labels = textos, shadow = True, autopct = '%1.1f%%')

fig.set_size_inches(10,10)
fig.show()
```
