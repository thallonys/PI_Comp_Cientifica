```python
import pandas as pd
import numpy as np
import math
```

1. importar os dados de CSV do dataset de seeds:

2. colocar as linhas de cabeçalho: 
  1. Área A,
  2. Perímetro P,
  3. Extensão do núcleo,
  4. Largura,
  5. Coeficiente de Assimetria
  6. Extensão do sulgo do núcleo.

```python
df = pd.read_csv('https://raw.githubusercontent.com/celsocrivelaro/simple-datasets/main/seeds.csv', header = None)
df.columns = ['Área A', 'Perímetro P', 'Extensão do núcleo', 'Largura', 'Coeficiente de Assimetria', 'Extensão do sulgo do núcleo', 'vazio1', 'vazio2', 'vazio3']
```

3. Remover colunas extras

```python
df.drop(labels = ["vazio1", "vazio2", "vazio3"], axis=1, inplace = True)
```

4. Remover linhas com valores nulos

```python
df.notna()
```

5. Adicionar um campo Compactação cujo o cálculo é $C = 4\pi\frac{A}{P^2}$

```python
def calcular_compactacao():
    c = 4 * np.pi * df['Área A'] / pow(df['Perímetro P'], 2)
    return c

df['Compactação'] = calcular_compactacao()

df
```

6. Exportar para CSV o valor final

```python
df.to_csv('exercício6_tratamento_de_dados_concluído.csv')
```
