## Tangente Hiperbólica

$ y = tanh(x) $

```python
import matplotlib.pyplot as plt
import numpy as np

amplitude_eixo_x = 10

x = np.linspace(- amplitude_eixo_x, amplitude_eixo_x, 100)
y = np.tanh(x)


fig, ax = plt.subplots()

fig.set_size_inches(20, 8)

ax.plot(x, y, label = "tanh(x)")
ax.grid()
ax.legend()
```

## Séries de Taylor aplicada à Tangente Hiperbólica

$ y = tanh(a) + \frac{tanh'(a)(x - a)^1}{1!} + \frac{tanh(a)''(x - a)^2}{2!} + \frac{tanh(a)'''(x - a)^3}{3!}\dots \frac{tanh^n(a)(x - a)^n}{n!}$

##### Definindo n (número de derivações) em $ tanh^n $ e considerando $ x = 0 $:

$ tanh^0(x) = 0 $

$ tanh^1(x) = sech^2(x) = 1 $

$ tanh^2(x) = - 2 tanh(x)sech^2(x) = 0 $

$ tanh^3(x) = -2sech^4(x) + 4sech^2(x)tanh^2(x)= -2$

$ tanh^4(x) = 16sech^4(x)tanh(x)-8sech^2(x)tanh^3(x) = 0 $

$ tanh^5(x) = 8sech^2(x)(2sech^4(x)-11sech^2(x)tanh^2(x)+2tanh^4(x)) = 16 $

$ tanh^6(x) = -16 (17 sech^6(x) tanh(x) - 26 sech^4(x) tanh^3(x) + 2 sech^2(x) tanh^5(x)) = 0 $

$ tanh^7(x) = -16 (17 sech^8(x) - 180 sech^6(x) tanh^2(x) + 114 sech^4(x) tanh^4(x) - 4 sech^2(x) tanh^6(x)) = -272 $

$ tanh^8(x) = -4(123 - 56cosh(2x) + cosh(4x))sech^6(x)tanh(x) = 0$


##### Então, eu tenho o polinômio:

$ tanh(x) = \frac{x^1}{1!} + \frac{0}{2!} - \frac{2x^3}{3!} + \frac{0}{4!} + \frac{16x^5}{5!} + \frac{0}{6!} - \frac{272x^7}{7!} \dots $

```python
import matplotlib.pyplot as plt
import numpy as np

amplitude_eixo_x = 1

x = np.linspace(- amplitude_eixo_x, amplitude_eixo_x, 100)
y1 = np.tanh(x)
y2 = x/1 - (2*x**3)/6 + (16*x**5)/120 - (272*x**7)/5040

fig, ax = plt.subplots()
fig.set_size_inches(20, 8)

ax.plot(x, y1, label = "tanh(x)")
ax.plot(x, y2, label = "tangenteHiperbolica")
ax.grid()
ax.legend()
```
