```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100) # sequência de 0 1 10
y = 3 + 2 * np.sin(2 * x)

# Essa função gera uma figura e eixos
fig, ax = plt.subplots()

# O ax são os eixos. Então, é ele que faz a plotagem (desenho)
ax.plot(x, y, "o") # plot imprime linhas ou pontos

fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100) # sequência de 0 1 10
y = 3 + 2 * np.sin(2 * x)

# Essa função gera uma figura e eixos
fig, axs = plt.subplots(1, 2) # 1 linha, 2 colunas - formato grid

# primeiro gráfico
axs[0].plot(x, y)

# segundo gráfico
axs[1].plot(x, y, "o")

# configura a largura e altura
fig.set_size_inches(20, 8)

fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100) # sequência de 0 1 10
y = 3 + 2 * np.sin(2 * x)

fig, axs = plt.subplots(1, 3)

axs[0].plot(x, y, color="green", linestyle="dashed")
axs[0].set_title("Gráfico 1")

axs[1].plot(x, y, color="blue", marker="o")
axs[1].set_title("Gráfico 2")

axs[2].plot(x, y, color="red", marker="s", linewidth=5, markersize= 10)
axs[2].set_title("Gráfico 3")

fig.set_size_inches(20, 8)

# Salva o gráfico como arquivo de imagem
fig.savefig("grafico.png")

fig.show()
```

Histograma

```python
import numpy as np
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_size_inches(20, 8)

# gaussina
# média, desvio
media, desvio = 50, 5
x = np.random.normal(media, desvio, 1000)
ax1.hist(x, edgecolor="white")
ax1.set_title("Gaussiana")
ax1.grid()
ax1.text(40, 220, r"$\mu=50,\sigma=5$")


# Poisson
eventos = 3
x = np.random.poisson(eventos, 1000)
ax2.hist(x, edgecolor="white")
ax2.set_title("Poisson")
ax2.grid()
ax2.text(4, 200, "Eventos = 3") # Insere texto pela coordenadas

fig.show()
```

Gráfico de Barras

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 0.5 + np.arange(8)
y = np.random.uniform(2, 7, len(x))

fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white")

ax.set(xlim=(0, 8), ylim=(0, 7)) # configura os espaçamentos das extremidades de x

fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 0.5 + np.arange(8)

valores = ["T1", "T2", "T3", "T4", "T5", "T6", "T7", "T8"]
y1 = np.array([1, 5, 7, 5, 2, 6, 8, 9])
y2 = np.array([9, 4, 3, 6, 7, 5, 3, 6])

fig, ax = plt.subplots()
largura_barra = 0.3
ax.bar(x - largura_barra/2, y1, width=largura_barra, label="Grupo 1")
ax.bar(x + largura_barra/2, y2, width=largura_barra, label="Grupo 2")

ax.set_xticks(x)
ax.set_xticklabels(valores)

ax.legend()

fig.show()
```

Gráfico de Pizza

```python
import numpy as np
import matplotlib.pyplot as plt

y = np.array([35, 25, 25, 15])
textos = ["Bananas", "Maçãs", "Pêras", "Uvas"]
explodir = [0.2, 0.1, 0, 0]

fig, ax = plt.subplots()

ax.pie(y, labels=textos, shadow=True, explode=explodir, autopct="%1.1f%%")

fig.show()
```
