Série de Taylor bobinha

```python
def cosseno_bobinho(x):
    return 1 - x*x/2 + x*x*x*x/(4*3*2) - x*x*x*x*x*x/(6*5*4*3*2)

trinta_graus = np.pi/6

print("Cosseno numpy__ 30º = ", np.cos(trinta_graus))
print("Cosseno bobinho 30º = ", cosseno_bobinho(trinta_graus))
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi, np.pi, 100)
y1 = np.cos(x)
y2 = cosseno_bobinho(x)

fig, ax = plt.subplots()
fig.set_size_inches(20, 8)
ax.plot(x, y1, linewidth = 2.0, label = "numpy")
ax.plot(x, y2, linewidth = 2.0, label = "bobinho")
ax.legend()
ax.grid()
```

```python
import math

def cosseno_melhor(x, n_max = 20):
    cosseno = 0.0
    for n in range(n_max):
        numerador = np.power(-1, n) * np.power(x, 2 * n)
        denominador = math.factorial(2 * n)
        cosseno = np.add(cosseno, numerador / denominador)
    return cosseno


print("Cosseno numpy   - 30º = ", np.cos(trinta_graus))
print("Cosseno bobinho - 30º = ", cosseno_bobinho(trinta_graus))
print("Cosseno melhor  - 30º = ", cosseno_melhor(trinta_graus))
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5 * np.pi, 5 * np.pi, 100)
y1 = np.cos(x)
y2 = cosseno_bobinho(x)
y3 = cosseno_melhor(x, 18)

fig, ax = plt.subplots()
fig.set_size_inches(20, 8)
ax.plot(x, y1, linewidth = 2.0, label = "numpy")
# ax.plot(x, y2, linewidth = 2.0, label = "bobinho")
ax.plot(x, y3, linewidth = 2.0, label = "melhor")
ax.legend()
ax.grid()
```

```python
import timeit

start = timeit.default_timer()
cosseno_bobinho(trinta_graus)
t_cos_bobinho = (timeit.default_timer() - start) * 1000

start = timeit.default_timer()
cosseno_melhor(trinta_graus)
t_cos_melhor = (timeit.default_timer() - start) * 1000

start = timeit.default_timer()
cosseno_melhor(trinta_graus)
t_cos_melhor = (timeit.default_timer() - start) * 1000

print("Cosseno bobinho: ", t_cos_bobinho)
print("Cosseno melhor: ", t_cos_melhor)

```

```python
import math

def cosseno_otimizado(x, n_max = 20):
    cosseno = 1.0
    numerador = -x * x
    denominador = 2.0
    cosseno = 1.0 + numerador / denominador

    for n in range(1, n_max):
        numerador = numerador * x * x
        denominador = denominador * (2 * n) * (2 * n - 1)

        cosseno += numerador / denominador
    return cosseno


print("Cosseno numpy      - 30º = ", np.cos(trinta_graus))
print("Cosseno bobinho    - 30º = ", cosseno_bobinho(trinta_graus))
print("Cosseno melhor     - 30º = ", cosseno_melhor(trinta_graus))
print("Cosseno otimizado  - 30º = ", cosseno_melhor(trinta_graus))
```

```python
import timeit

start = timeit.default_timer()
cosseno_bobinho(trinta_graus)
t_cos_bobinho = (timeit.default_timer() - start) * 1000

start = timeit.default_timer()
cosseno_melhor(trinta_graus)
t_cos_melhor = (timeit.default_timer() - start) * 1000

start = timeit.default_timer()
cosseno_otimizado(trinta_graus)
t_cos_otimizado = (timeit.default_timer() - start) * 1000

print("Cosseno bobinho: ", t_cos_bobinho)
print("Cosseno melhor: ", t_cos_melhor)
print("Cosseno otimizado: ", t_cos_otimizado)
```
